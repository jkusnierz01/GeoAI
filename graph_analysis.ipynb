```python
import torch
from torch_geometric.utils import to_networkx
import matplotlib.pyplot as plt
import networkx as nx

data = torch.load("/mnt/data/kusnierz/geo/GeoAI/dataset_aligned/los_angeles_hexagons_res7.pt", map_location="cpu", weights_only=False)
```

```python
# Jeśli edge_index ma więcej niż 2 wiersze, weź tylko pierwsze dwa
if data.edge_index.shape[0] > 2:
    print(f"edge_index shape before: {data.edge_index.shape}")
    data.edge_index = data.edge_index[:2]
    print(f"edge_index shape after: {data.edge_index.shape}")

G = to_networkx(data, to_undirected=True)

plt.figure(figsize=(8, 8))
nx.draw(G, node_size=20, with_labels=False)
plt.show()
```

```python
is_connected = nx.is_connected(G)
print("Graf spójny:", is_connected)
```
